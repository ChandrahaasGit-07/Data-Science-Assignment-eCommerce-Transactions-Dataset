```python
# Step 2 :   Univariate Analysis (Single Variables)
```

```python
quality_counts = df['categer'].value_counts()

plt.figure(figsize=(8, 6))
plt.bar(quality_counts.index, quality_counts, color='red')
plt.title('Count Plot of Quality')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.show()
```

```python
# Step 3: Bivariate Analysis for (understanding variable interactions and correlations effectively)
```

```python
# Bivariate Analysis: Region vs Sales (replace 'Sales' with the actual numerical column you're analyzing)

plt.figure(figsize=(8, 6))

# Boxplot to visualize the relationship between 'Region' and 'Sales'
sns.boxplot(x='Category', y='Price', data=df, palette='Set2')

# Add title and labels
plt.title('Category vs Price')
plt.xlabel('Category')
plt.ylabel('Price')

plt.show()

# Alternatively, use a Violin plot for better distribution visualization
plt.figure(figsize=(8, 6))

sns.violinplot(x='Category', y='Price', data=df, palette='Set2')

# Add title and labels
plt.title('Category vs Price (Violin Plot)')
plt.xlabel('Category')
plt.ylabel('Price')

plt.show()
```
